```python
import matplotlib.pyplot as plt
import numpy as np
```

Linear Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.plot(x,y)
plt.show()

```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = 'r'  #color change
plt.plot(x,y,c)
plt.show()
```

```python
#random values
x = [10,5,6,9,8,7,4]
y =[4,5,6,9,8,7,2]

plt.plot(x,y)
plt.show()
```

```python
#using marker

x = [10,5,6,9,8,7,4]
y =[4,5,6,9,8,7,2]

plt.plot(x,y, marker='o')
plt.show()
```

```python
#using marker in only y axis
x = [10,5,6,9,8,7,4]
y =[4,5,6,9,8,7,2]

plt.plot(x,y)
plt.plot(y, marker= ">")
plt.show()
```

```python
#using marker in only X axis
x = [10,5,6,9,8,7,4]
y =[4,5,6,9,8,7,2]

plt.plot(x,y)
plt.plot(x, marker= "s")
plt.show()
```

```python
#using marker in only y axis and X axis both
x = [10,5,6,9,8,7,4]
y =[4,5,6,9,8,7,2]

plt.plot(x,y)
plt.plot(y, marker= ">")
plt.plot(x, marker= "*")
plt.show()
```

Bar Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
#color bar graph
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','y','b']
plt.bar(x,y,color = c)
plt.show()
```

```python
x = np.array (["a","b","c","d"])
y= np.array([3,8,5,6])

plt.barh(x,y)  # using for horizontal bars
plt.show()
```

```python
#same color for all the bar
x = np.array (["a","b","c","d"])
y= np.array([3,8,5,6])

plt.bar(x,y, color= "pink")  
plt.show()
```

```python
# adjust the width 
x = np.array (["a","b","c","d"])
y= np.array([3,8,5,6])

plt.bar(x,y, width = 0.1)  
plt.show()
```

```python
#height 
x = np.array (["a","b","c","d"])
y= np.array([3,8,5,6])

plt.barh(x,y ,height =0.1)
plt.show()
```

Scatter Plot

```python
x = (1,2,3,4)
y= (3,8,5,6)
plt.scatter(x,y)  
plt.show()
```

```python
# give the title
x = (1,2,3,4)
y= (3,8,5,6)
plt.title("scatter plot")
plt.scatter(x,y)  
plt.show()
```

```python
# adding the label on X axis
x = (1,2,3,4)
y= (3,8,5,6)
plt.title("scatter plot")
plt.xlabel("month")
plt.scatter(x,y)  
plt.show()
```

```python
# adding the label on y axis
x = (1,2,3,4)
y= (3,8,5,6)
plt.title("scatter plot")
plt.ylabel("amount")
plt.scatter(x,y)  
plt.show()
```

```python
# adding the label on y axis and x axis
x = (1,2,3,4)
y= (3,8,5,6)
plt.title("scatter plot")
plt.ylabel("amount")
plt.xlabel("month")
plt.scatter(x,y)  
plt.show()
```

```python
# change the color of scatter plot
x = (1,2,3,4)
y= (3,8,5,6)
plt.title("scatter plot")
plt.ylabel("amount")
plt.xlabel("month")

c = ["red" ,"yellow","pink", "blue"]
plt.scatter(x,y, color = c)  
plt.show()
```

```python
# Size
x = (1,2,3,4)
y= (3,8,5,6)
plt.title("scatter plot")
plt.ylabel("amount")
plt.xlabel("month")
plt.scatter(x,y , s=150)  
plt.show()
```
